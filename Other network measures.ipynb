# Setup

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a dict of edges
fname3 = 'celegans_connectome_multiplex.edges'
print(f'reading the network file : {fname3}\n')
f = open('./' + fname3, 'r')

edges1 = {}
edges2 = {}
edges3 = {}
counter1 = 0
counter2 = 0
counter3 = 0

for line in f:
    layer, node1, node2, w = map(int, line.strip().split(' '))
    if layer == 1:
        edges1[counter1] = node1, node2
        counter1 += 1
        #print(1)
    if layer == 2:
        edges2[counter2] = node1, node2
        counter2 += 1
        #print(2)
    if layer == 3:
        edges3[counter3] = node1, node2
        counter3 += 1
        #print(3)

# create new edges from the dictionary
new_edges1 = []
for idx, e in edges1.items():
    new_edges1.append(e)
new_edges2 = []
for idx, e in edges2.items():
    new_edges2.append(e)
new_edges3 = []
for idx, e in edges3.items():
    new_edges3.append(e)
G1 = nx.MultiGraph(new_edges1)
G2 = nx.MultiGraph(new_edges2)
G3 = nx.MultiGraph(new_edges3)
print(G1)
print(G2)
print(G3)
```

## Multiedge nodes

```python
def multiedge_nodes(G, e):
    multi_e = []
    sum_multi_e = 0
    for u, v in G.edges():
        if G.number_of_edges(u, v) > 1:
            if (u,v) not in multi_e:
                multi_e.append((u,v))
                sum_multi_e += G.number_of_edges(u, v)     
    for e in multi_e:
        print(e)
    print(sum_multi_e)
```

```python
print("Layer 1 multiedges")
multiedge_nodes(G1, edges1) # 1028
print("Layer 2 multiedges:")
multiedge_nodes(G2, edges2)  # 1502
print("Layer 3 multiedges:")
multiedge_nodes(G3, edges3)  # 2980
```

##  Specific node labels and ids in similar layers

```python
def selfloops(G):
    edges_with_selfloops = list(nx.selfloop_edges(G))
    nodes_with_selfloops = list(nx.nodes_with_selfloops(G))
    print("Edges related to self-loop edges:", edges_with_selfloops)
    print("Nodes related to self-loop nodes:", nodes_with_selfloops)
    return nodes_with_selfloops
```

```python
n1 = selfloops(G1)
print(n1)
n2 = selfloops(G2)
print(n2)
n3 = selfloops(G3)
print(n3)
```

```python
def self_loop_node_label(nodeSL, nodeID, nodeLabel):
    nL = ""
    for n in nodeSL:
        if int(nodeID) == int(n):
            nL = nodeLabel
    return(nL)
```

```python
fname2 = 'celegans_connectome_nodes.txt'
print(f'reading the network file : {fname2}\n')
f = open('./' + fname2, 'r')
overlapping_n = []
overlapping_n2 = []
overlapping_n3 = []
line_count = 0
for line in f:
    if line_count != 0:
        ID, label = line.strip().split(' ')
        nL = self_loop_node_label(n1, ID, label)
        nL2 = self_loop_node_label(n2, ID, label)
        nL3 = self_loop_node_label(n3, ID, label)
        if nL != "":
            overlapping_n.append(nL)
        if nL2 != "":
            overlapping_n2.append(nL2)
        if nL3 != "":
            overlapping_n3.append(nL3)
    line_count = line_count + 1
print(overlapping_n)
print(overlapping_n2)
print(overlapping_n3)
```

## Connected components

```python
h1 = list(nx.connected_components(G1))
LCC1 = max(h1, key=len)
print("h1", len(h1))
print("LCC1", len(LCC1))
h2 = list(nx.connected_components(G2))
LCC2 = max(h2, key=len)
print("h2", len(h2))
print("LCC2", len(LCC2))
h3 = list(nx.connected_components(G3))
LCC3 = max(h3, key=len)
print("h3", len(h3))
print("LCC1", len(LCC3))
```

## Highest degree and node label

```python
most_connected_node1 = max(G1.nodes(), key=G1.degree)
max_degree = G1.degree(most_connected_node1)
print("Most connected node:", most_connected_node1)
print("Degree of the most connected node:", max_degree)

most_connected_node2 = max(G2.nodes(), key=G2.degree)
max_degree = G2.degree(most_connected_node2)
print("Most connected node:", most_connected_node2)
print("Degree of the most connected node:", max_degree)

most_connected_node3 = max(G3.nodes(), key=G3.degree)
max_degree = G3.degree(most_connected_node3)
print("Most connected node:", most_connected_node3)
print("Degree of the most connected node:", max_degree)

fname2 = 'celegans_connectome_nodes.txt'
print(f'reading the network file : {fname2}\n')
f = open('./' + fname2, 'r')
nodeLabel = []
line_count = 0
for line in f:
    if line_count != 0:
        ID, label = line.strip().split(' ')
        if int(ID) == int(most_connected_node1):
            nodeLabel.append(label)
        if int(ID) == int(most_connected_node2):
            nodeLabel.append(label)
        if int(ID) == int(most_connected_node3):
            nodeLabel.append(label)
    line_count = line_count + 1
print(nodeLabel)
```

## Degree centrality and node label

```python
deg_c = nx.degree_centrality(G1)
n1 = max(G1.nodes(), key=deg_c.get)
max_degree1 = deg_c[n1]
print("Node with highest degree centrality:", n1)
print("Degree centrality of the node:", max_degree1)

deg_c = nx.degree_centrality(G2)
n2 = max(G2.nodes(), key=deg_c.get)
max_degree2 = deg_c[n2]
print("Node with highest degree centrality:", n2)
print("Degree centrality of the node:", max_degree2)

deg_c = nx.degree_centrality(G3)
n3 = max(G3.nodes(), key=deg_c.get)
max_degree3 = deg_c[n3]
print("Node with highest degree centrality:", n3)
print("Degree centrality of the node:", max_degree3)

fname2 = 'celegans_connectome_nodes.txt'
print(f'reading the network file : {fname2}\n')
f = open('./' + fname2, 'r')
nodeLabel = []
line_count = 0
for line in f:
    if line_count != 0:
        ID, label = line.strip().split(' ')
        if int(ID) == int(n1):
            nodeLabel.append(label)
        if int(ID) == int(n2):
            nodeLabel.append(label)
        if int(ID) == int(n3):
            nodeLabel.append(label)
    line_count = line_count + 1
print(nodeLabel)
```
