# Configuration Model

Nomenclature:
* G1 is the graph of all edges in layer 1 (G2 for layer 2, G3 for layer 3)
* Similarly, config_model_1 is the configuration model of layer 1 (2 for layer 2, 3 for layer 3)

```python
#Configuration Model Function layout (will need editing)

def create_and_visualize_configuration_model(G, title):
    # Create a list of degrees for each node
    degree_sequence = [d for n, d in G.degree()]

    # Generate a configuration model graph based on the degree sequence
    config_model = nx.configuration_model(degree_sequence, create_using=nx.MultiGraph())

    # Visualize the configuration model graph
    pos = nx.spring_layout(config_model)  # Layout for visualization
    nx.draw(config_model, pos, with_labels=False, node_color='lightseagreen', node_size=30, edge_color='gray')
    plt.title(title)
    plt.show()
    return(config_model)
```

```python
# Apply function (def a ridiculogram...)
config_model_1= create_and_visualize_configuration_model(G1, "Configuration Model 1")
print(config_model_1)
```

```python
config_model_2 = create_and_visualize_configuration_model(G2, "Configuration Model 2")
print(config_model_2)
```

```python
config_model_3 = create_and_visualize_configuration_model(G3, "Configuration Model 3")
print(config_model_3)
```

```python
#Visualization of actual connectome
def elegans_concectome_visualization(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=False, node_color='orangered', node_size=30, edge_color='gray')
    plt.title(title)
    plt.show()
```

```python
def analyze_connectome_and_configuration_model(original_graph, config_model):
    # Calculate degree distribution
    celegans_degrees = [d for n, d in original_graph.degree()]
    config_degrees = [d for n, d in config_model.degree()]

    # Print results
    print("Degree Distribution:")
    print("C. elegans Connectome:", celegans_degrees)
    print("Configuration Model:", config_degrees)
```

```python
# Function to compute clustering coefficient for a node in a multigraph
def multigraph_clustering_coefficient(G, node):
    neighbors = list(G.neighbors(node))
    if len(neighbors) < 2:
        return 0.0  # If there are fewer than 2 neighbors, clustering coefficient is 0
    # Count the number of edges between neighbors
    num_edges_between_neighbors = sum(G.number_of_edges(neighbor1, neighbor2) for i, neighbor1 in enumerate(neighbors) for neighbor2 in neighbors[i+1:])
    # Compute clustering coefficient
    clustering_coefficient = 2 * num_edges_between_neighbors / (len(neighbors) * (len(neighbors) - 1))
    return clustering_coefficient


#Compute clustering coefficients for each node
node_clustering = {node: multigraph_clustering_coefficient(G3, node) for node in G3.nodes()}

#Compute average clustering coefficient
total_clustering_coefficient = sum(node_clustering.values())
average_clustering = total_clustering_coefficient / len(G3)

Print the clustering coefficients
print("Node clustering coefficients:", node_clustering)
print("Average clustering coefficient:", average_clustering)
```
